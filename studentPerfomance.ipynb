```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('StudentPerformanceFactors.csv')
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.dropna(subset = ['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home'], inplace = True)
data.isna().sum()
```

```python
data.describe()
```

```python
rows_exceeding_100 = data[data['Exam_Score'] > 100]
rows_exceeding_100
```

```python
data.loc[data['Exam_Score'] > 100, 'Exam_Score'] = 100
data.describe()
```

```python
data.head()
```

```python
avg_score_by_gender = data.groupby('Gender')['Exam_Score'].mean()
avg_score_by_gender
```

```python
avg_score_by_gender.plot(kind='bar', color = 'green')
plt.title('Average Exam Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.show()
```

```python
plt.scatter(data['Hours_Studied'], data['Exam_Score'])
plt.title('Hours Studied vs Score')
plt.xlabel('Hours Studied')
plt.ylabel('Avg Score')
plt.show()
```

```python
high_score_low_hours = data[(data['Hours_Studied'] <= 1) & (data['Exam_Score'] >=90)]
high_score_low_hours
```

```python
high_hour_studies = data[(data['Hours_Studied'] >=35)]
high_hour_studies.describe()
```

```python
avg_score_by_parental_involvement = data.groupby('Parental_Involvement')['Exam_Score'].mean()
colors = ['blue', 'green', 'red']
avg_score_by_parental_involvement.plot(kind='bar', color = colors)
plt.title('Average Exam Score by Parental Involvement')
plt.xlabel('Parental Involvement')
plt.ylabel('Average Exam Score')
plt.show()
avg_score_by_parental_involvement
```

```python
avg_score_by_attendace = data.groupby('Attendance')['Exam_Score'].mean()
avg_score_by_attendace.plot(kind='line', color='orange')
plt.title('Average Exam Score vs Attendance')
plt.xlabel('Attendance')
plt.ylabel('Average Exam Score')
plt.show()
```
